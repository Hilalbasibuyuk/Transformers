```python
from transformers import AutoTokenizer
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
encoding = tokenizer("I like this so much!")
print(encoding)
```

```python
pt_batch = tokenizer(
    ["I like this so much!"],
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="pt", # Pytorch tensor
)
```

```python
print(pt_batch)
```

```python
tf_batch = tokenizer(
    ["I like this so much"],
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="tf", # Tensorflow tensor
)
```

```python
print(tf_batch)
```
