```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
classifier("I am not happy to work Transformers.")
```

```python
classifier("I am happy to work Transformers.")
```

```python
results = classifier(["You don't understand me", "We hope you don't hate it.", "This is a banana"])
for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```
